```python
import numpy as np
```

```python
n=200
d=2
```

```python
X=np.random.randn(n,d)
```

```python
X
```

```python
ones=np.array([[1]*n]).T
```

```python
ones
```

```python
Xb=np.concatenate((ones,X),axis=1)
```

```python
Xb
```

```python
w=np.random.randn(d+1)
```

```python
z=Xb.dot(w)
```
