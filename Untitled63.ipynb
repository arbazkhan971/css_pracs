```python
from PIL import Image
import matplotlib.pyplot as plt 
import numpy as np 
 
img1=Image.open("IMG_20180408_005214.jpg") 
img2=Image.open("IMG_20180408_005245.jpg") 
 
#Image in array format 
arrayimg1=np.array(img1) 
arrayimg2=np.array(img2) 
 
fig=plt.figure() 
 
# Face I 
ax1=fig.add_subplot(2,2,1)
ax1.title.set_text("Face I") 
ax1.set_xticklabels([])
ax1.set_yticklabels([]) 
imgplot=plt.imshow(img1) 
 
# Face II 
ax2=fig.add_subplot(2,2,2)
ax2.title.set_text("Face II") 
ax2.set_xticklabels([])
ax2.set_yticklabels([]) 
imgplot=plt.imshow(img2) 
 
# Linear combination of Face I & Face II
a=1
b=2 
linearcom=(a*arrayimg2)+(b*arrayimg2)
linecomimg=Image.fromarray(linearcom) 
ax3=fig.add_subplot(2,2,3) 
ax3.title.set_text("Linear Combination of Face I & Face II") 
ax3.set_xticklabels([]) 
ax3.set_yticklabels([]) 
imgplot=plt.imshow(linecomimg) 
plt.show()
# Average face of original faces
r=arrayimg1+arrayimg2 
avg=r//2 
avgimg=Image.fromarray(avg) 
ax4=fig.add_subplot(2,2,4) 
ax4.title.set_text("Average Face of Face I & Face II")
ax4.set_xticklabels([])
ax4.set_yticklabels([]) 
imgplot=plt.imshow(avgimg) 
plt.show() 

```
