```python
N=Function("N")
t=Symbol("t")
r=Symbol("r")
k=Symbol("k")
```

```python
N(t)
```

```python
ode=Eq(N(t).diff(t),r*N(t)*(1-N(t)/k))
```

```python
Solution=dsolve(ode,N(t))
```

```python
Solution
```

```python
general_solution = sympy.dsolve(ode, N(t))
value = general_solution.subs([(t,3), (N(0), 1)])
value
```

```python
ode_solution = general_solution.subs([(value.rhs,value.lhs)])
ode_solution
```
