```python
from selenium import webdriver
from bs4 import BeautifulSoup

driver = webdriver.PhantomJS(executable_path = r'C:\Users\arbaz\Downloads\phantomjs-2.1.1-windows\phantomjs-2.1.1-windows\bin\phantomjs.exe')
```

```python
driver.get('https://www.consumerreports.org/cro/a-to-z-index/products/index.htm')
```

```python
html_doc = driver.page_source
```

```python
soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
class_div=soup.find_all('div',{"class":"crux-body-copy"})
```

```python
prod_name=[]
prod_link=[]
for i in class_div:
    prod_name.append(i.a.text)
    prod_link.append(i.a["href"])
```
